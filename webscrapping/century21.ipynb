```python
import requests
from bs4 import BeautifulSoup

r=requests.get('https://www.century21.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/')
c=r.content

soup=BeautifulSoup(c,"html.parser")

all=soup.find_all("div",{"class":"property-card-primary-info"})

all[0].find("a",{"class":"listing-price"}).text.replace("\n","").replace(" ","")
```

```python
l=[]
for item in all:
    d={}
    d["Price"]=item.find("a",{"class":"listing-price"}).text.replace("\n","").replace(" ","")
    d["Beds"]=item.find_all("div",{"class","property-beds"})[0].text.replace("\n","").replace(" ","")
    d["Baths"]=item.find_all("div",{"class","property-baths"})[0].text.replace("\n","").replace(" ","")
    try:
        d["Half-baths"]=item.find_all("div",{"class","property-half-baths"})[0].text.replace("\n","").replace(" ","")
    except:
        d["Half-baths"]=None
    try:
        d["Area"]=item.find_all("div",{"class","property-sqft"})[0].text.replace("\n","").replace(" ","")
    except:
        d["Area"]=None
    d["Address"]=item.find_all("div",{"class","property-address"})[0].text.replace("\n","").replace(" ","")
    try:
        d["Location"]=item.find_all("div",{"class","property-city"})[0].text.replace("\n","").replace(" ","")
    except:
        d["Location"]=None
    
    try:
        d["Attribution"]=item.find_all("div",{"class","property-card-attribution"})[0].text.replace("\n","").replace(" ","")
    except:
        d["Attribution"]=None
    l.append(d)
```

```python
import pandas 
df=pandas.DataFrame(l)
```
